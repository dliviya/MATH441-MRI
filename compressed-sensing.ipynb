```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
import scipy.linalg as la
import cv2
```

```python
# open the image 
image = Image.open("/Users/elisegreen/Downloads/dog.jpg")
#coverting the image to black and white so it is easier to process
image = image.convert('L')
```

```python
array = np.array(image)

height = array.shape[0]
width = array.shape[1]
```

```python
#number of pixels we are going to take, in this case we are starting with 10%
n = int(0.1*width*height)

# randomly pick the locations for the pixels which we will use
x_coords = np.random.randint(low = 0, high = 360, size = n)
y_coords = np.random.randint(low = 0, high = 409, size = n)


pixel_grid = np.zeros((361,410))


for m  in range(n):
    x = x_coords[m]
    y = y_coords[m]
    pixel_grid[x][y] = 1
```

```python
sub_sample = array*pixel_grid
plt.imshow(Image.fromarray(sub_sample))
```

```python
list = []
for x in range(height):
    for y in range(width):
        if sub_sample[x][y] != 0:
            list.append(sub_sample[x][y])

y = np.array(list)
```
